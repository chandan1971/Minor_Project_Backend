```python
# Here, we will do below things:
# 1 . convert tags column into a vector by using CountVectorizer .
# 2 . Reduce similar words into their root origin word by using PorterStemmer .

import sklearn

from sklearn.feature_extraction.text import CountVectorizer
import nltk
from nltk.stem.porter import PorterStemmer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
vectors = cv.fit_transform(new_df['tags']).toarray()
ps = PorterStemmer()
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return ' '.join(y)
```
